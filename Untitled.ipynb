```python
import numpy as np
import pandas as pd
```

```python
data_train = 'data/DailyDelhiClimateTrain.csv'
data_test = 'data/DailyDelhiClimateTest.csv'
df_train = pd.read_csv(data_train)
df_test = pd.read_csv(data_test)
```

```python
df_test['date'] = pd.to_datetime(df_test['date']) 
df_test['date'] = (df_test['date'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
df_test.head()
```

```python
df_train['date'] = pd.to_datetime(df_train['date']) 
df_train['date'] = (df_train['date'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
df_train.head()
```

```python
def returnLegValue(column_name,column,df,num_of_variables):
    j=0
    pom=1
    for j in range(num_of_variables):
        col = column_name + str(j)
        for i in range(df.shape[0]-pom):
            df.loc[i+pom,col] = df.loc[i+pom,column] - df.loc[i,column]
        pom += 1
```

```python
returnLegValue('temp_diference_','meantemp',df_train,4)
returnLegValue('wind_speed_diference_','wind_speed',df_train,4)
returnLegValue('meanpressure_diference_','meanpressure',df_train,4)
df_train.head()
```

```python
returnLegValue('temp_diference_','meantemp',df_test,4)
returnLegValue('wind_speed_diference_','wind_speed',df_test,4)
returnLegValue('meanpressure_diference_','meanpressure',df_test,4)
df_test.head()
```

```python
df_train = df_train.fillna(df_train.mean())
df_train.head()
```

```python
df_test = df_test.fillna(df_test.mean())
df_test.head()
```

```python
target = df_test.pop('meantemp')
df_test['meantempt'] = target
df_test.head()
```

```python
target = df_train.pop('meantemp')
df_train['meantempt'] = target
df_train.head()
```

```python
df_train.to_csv('data/train_data.csv', index=False)
df_test.to_csv('data/test_data.csv', index=False)
```
